```python
ГРУППИРОВАТЬ ПО
Функция groupby() в Pandas используется для группировки данных по одному или нескольким столбцам в DataFrame.
```

```python
df.groupby(['Size'], as_index=False)['Qty'].sum().sort_values(by='Qty',ascending=False)
```

```python
# Проверка данных B2B с помощью круговой диаграммы
B2B_Check = df['B2B'].value_counts()

#  Plot the pie chart
plt.pie(B2B_Check, labels=B2B_Check, autopct='%1.1f%%')
#plt.axis('equal')
plt.show()
```

```python
# Проверка данных B2B с помощью круговой диаграммы

B2B_Check = df['B2B'].value_counts()

#  Plot the pie chart
plt.pie(B2B_Check, labels=B2B_Check.index, autopct='%1.1f%%')
#plt.axis('equal')
plt.show()
```

```python
*Примечание: из приведенной выше диаграммы видно, что максимум 99,2% покупателей — это розничные торговцы, а 0,8% — покупатели B2B.
```

```python
#  Подготовка данных для круговой диаграммы

a1 = df['Fulfilment'].value_counts()

# Step 4: Plot the pie chart
fig, ax = plt.subplots()

ax.pie(a1, labels=a1.index, autopct='%1.1f%%', radius=0.7, wedgeprops=dict(width=0.6))
ax.set(aspect="equal")

plt.show()
```

```python
*Примечание: Из приведенной выше диаграммы вы можете видеть, что большая часть заказов выполняется через Amazon.
```

```python
# Подготовка данных для построения диаграммы рассеяния

x_data = df['Category']  
y_data = df['Size'] 

# Plot the scatter plot
plt.scatter(x_data, y_data)
plt.xlabel('Category ')  
plt.ylabel('Size')  
plt.title('Scatter Plot') 
plt.show()
```

```python
# Построить график количества городов по штатам

plt.figure(figsize=(12, 6))
sns.countplot(data=df, x='ship-state')
plt.xlabel('ship-state')
plt.ylabel('count')
plt.title('Distribution of State')
plt.xticks(rotation=90)
plt.show()
```

```python
# ТОП 10 ШТАТОВ

top_10_state = df['ship-state'].value_counts().head(10)
# Plot count of cities by state
plt.figure(figsize=(12, 6))
sns.countplot(data=df[df['ship-state'].isin(top_10_state.index)], x='ship-state')
plt.xlabel('ship-state')
plt.ylabel('count')
plt.title('Distribution of  State')
plt.xticks(rotation=45)
plt.show()
```

```python
*Примечание: на графике выше видно, что большинство покупателей — жители штата Махараштра.
```

```python
*Анализ данных показывает, что компания имеет значительную клиентскую базу в штате Махараштра, обслуживает в основном розничных продавцов,
выполняет заказы через Amazon,испытывает высокий спрос на футболки и считает размер M предпочтительным выбором среди покупателей.
```
